```python
corrmat = df.corr()
f, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(corrmat, ax=ax, cmap = 'RdYlBu_r', linewidths = 0.6)
plt.show()
```

```python
plt.figure(figsize=(50,11))
plt.bar(df['Region'], df[' Estimated Employed'])
plt.title('Region vs Estimated_Employed')
plt.xlabel('Region')
plt.ylabel('Estimated_Employed')
plt.show()
```

```python
plt.figure(figsize=(53,11))
plt.bar(df['Region'], df[' Estimated Labour Participation Rate (%)'])
plt.title('Region vs Estimated Labour Participation Rate (%)')
plt.xlabel('Region')
plt.ylabel('Estimated Labour Participation Rate (%)')
plt.show()
```

```python
plt.figure(figsize=(50,11))
plt.bar(df['Region'], df[' Estimated Unemployment Rate (%)'])
plt.title('Region vs Estimated_Unemployment_Rate(%')
plt.xlabel('Region')
plt.ylabel('Estimated_Unemployment_rate(%)')
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.histplot(df[' Estimated Unemployment Rate (%)'], bins =100, kde=True, palette='hls')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.histplot(df[' Estimated Employed'], kde=True, bins=100, palette='hls')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.histplot(df[' Estimated Labour Participation Rate (%)'], bins=100, palette='hls', kde=True)
plt.xticks(rotation=90)
plt.show()
```
